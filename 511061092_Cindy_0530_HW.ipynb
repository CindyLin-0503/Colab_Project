```python
import nltk
from nltk.corpus import PlaintextCorpusReader

corpus_root = "/content/drive/MyDrive/Colab Notebooks/Scripts"
wordlists = PlaintextCorpusReader(corpus_root, ".*")

book = wordlists.fileids()
print(book)
```

```python
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:7])
    for fileid in wordlists.fileids()
    for w in wordlists.words(fileid)
    for target in ["the","we","people"]
    if w.lower() == target)
cfd.plot()
```

```python
import nltk
from wordcloud import STOPWORDS

dataset = wordlists.raw('Oprah Winfreyâ€™s Golden Globes Speech.txt')

tokens = dataset.split(' ')
clean_tokens = tokens[:]

for token in tokens:
   if token.lower() in STOPWORDS:
    clean_tokens.remove(token)

freq = nltk.FreqDist(clean_tokens)

freq.plot(20, cumulative=False)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/Colab Notebooks/Scripts/Oprah Winfreyâ€™s Golden Globes Speech.txt',"r").read()

stop=["people","old people","time","life"] + list(STOPWORDS)

wordcloud = WordCloud(
    background_color="white",
    width=1500,
    height=1000,
    margin= 2,
    stopwords=stop
).generate(f)

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
